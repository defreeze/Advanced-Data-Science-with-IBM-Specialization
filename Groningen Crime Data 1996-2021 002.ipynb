```python
# import libraries
import pandas as pd
import numpy as np
from sklearn import preprocessing
```

```python
# load dataset
csvFile = pd.read_csv("aangifte-politie_totaal.csv")
df = pd.DataFrame(csvFile)
df.head()
```

```python
# data cleaning
np.where(pd.isnull(df))
```

```python
df.describe()
```

```python
df_transpose = df.transpose()
df_transpose = df_transpose.drop([df_transpose.index[0],df_transpose.index[1], df_transpose.index[3]])
df_transpose.head()
```

```python
headers = df_transpose.iloc[0].values
df_transpose.columns = headers
df_transpose = df_transpose.drop(df_transpose.index[0])
df_transpose.head()
```

```python
# simplify the transposed dataset
df_transpose = df_transpose.reset_index()
df_transpose = df_transpose.join(df_transpose['index'].str.split('_', 1, expand=True).rename(columns={0:'Year', 1:'Jaar'}))
df_transpose.drop(columns=['Year', 'index'], axis=1, inplace=True)
first_column = df_transpose.pop('Jaar')
df_transpose.insert(0, 'Jaar', first_column)
```

```python
df_transpose.head()
```

```python
# normalize the transposed data set
s0 = df_transpose.iloc[:,1:]
df_transpose.iloc[:,1:] = (s0 - s0.mean()) / (s0.max() - s0.min())
df_normalized = df_transpose
df_normalized.head()
```
