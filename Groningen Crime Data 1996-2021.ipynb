```python
# import libraries
import pandas as pd
import numpy as np
from sklearn import preprocessing
```

```python
# load dataset
csvFile = pd.read_csv("aangifte-politie_totaal.csv")
df = pd.DataFrame(csvFile)
df.head()
```

```python
# normalize the dataset
df = df.drop(['main_group', 'subgroup'], axis=1)
s0 = df.iloc[:,2:]
df.iloc[:,2:] = (s0 - s0.mean()) / (s0.max() - s0.min())
df
```
